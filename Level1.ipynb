```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("Dataset .csv")
sns.set_style("whitegrid")
```

```python
#TaskOne
cuisine_counts = df['Cuisines'].dropna().str.split(',').explode().str.strip().value_counts()
top_3_cuisines = cuisine_counts.head(3)

print(" Top 3 Most Common Cuisines:\n")
for cuisine, count in top_3_cuisines.items():
    print(f"{cuisine}: {count} restaurants")

top_3_percent = (
    df['Cuisines']
    .dropna()
    .apply(lambda x: any(cuisine in x for cuisine in top_3_cuisines.index))
    .value_counts(normalize=True) * 100
)

print("\n Percentage of restaurants serving any of the top 3 cuisines:")
print(top_3_percent)
```

```python
plt.figure(figsize=(6,4))
sns.barplot(x=top_3_cuisines.index, y=top_3_cuisines.values, palette="viridis")
plt.title("Top 3 Most Common Cuisines")
plt.ylabel("Number of Restaurants")
plt.show()
```

```python
#Task_two
city_counts = df['City'].value_counts()
top_city = city_counts.idxmax()

print(f"\n City with highest number of restaurants: {top_city} ({city_counts.max()} restaurants)")

avg_rating_city = df.groupby('City')['Aggregate rating'].mean().sort_values(ascending=False)
top_avg_rating_city = avg_rating_city.idxmax()

print("\n Average Rating per City (Top 10):")
print(avg_rating_city.head(10))

print(f"\n City with highest average rating: {top_avg_rating_city} ({avg_rating_city.max():.2f})")
```

```python
fig, axes = plt.subplots(1, 2, figsize=(14,5))

sns.barplot(x=city_counts.head(10).values, y=city_counts.head(10).index, palette="coolwarm", ax=axes[0])
axes[0].set_title("Top 10 Cities by Number of Restaurants")
axes[0].set_xlabel("Number of Restaurants")
axes[0].set_ylabel("City")

sns.barplot(x=avg_rating_city.head(10).values, y=avg_rating_city.head(10).index, palette="magma", ax=axes[1])
axes[1].set_title("Top 10 Cities by Avg Rating")
axes[1].set_xlabel("Average Rating")
axes[1].set_ylabel("City")

plt.tight_layout()
plt.show()
```

```python
#Task_Three
price_counts = df['Price range'].value_counts().sort_index()
price_percentage = (price_counts / len(df)) * 100

print("\n Percentage of restaurants in each price range:")
print(price_percentage)

plt.figure(figsize=(6,4))
sns.barplot(x=price_counts.index, y=price_counts.values, palette="Set2")
plt.title("Distribution of Price Ranges")
plt.xlabel("Price Range")
plt.ylabel("Number of Restaurants")
plt.show()
```

```python
#Task_Four
online_delivery_percentage = df['Has Online delivery'].value_counts(normalize=True) * 100
avg_rating_online = df.groupby('Has Online delivery')['Aggregate rating'].mean()

print("\n Percentage of restaurants offering online delivery:")
print(online_delivery_percentage)

print("\nAverage ratings by online delivery availability:")
print(avg_rating_online)
```

```python
fig, axes = plt.subplots(1, 2, figsize=(12,5))

sns.barplot(x=online_delivery_percentage.index, y=online_delivery_percentage.values, palette="pastel", ax=axes[0])
axes[0].set_title("Online Delivery Availability (%)")
axes[0].set_xlabel("Has Online Delivery")
axes[0].set_ylabel("Percentage")

sns.barplot(x=avg_rating_online.index, y=avg_rating_online.values, palette="autumn", ax=axes[1])
axes[1].set_title("Average Ratings: Online Delivery vs No Delivery")
axes[1].set_xlabel("Has Online Delivery")
axes[1].set_ylabel("Average Rating")

plt.tight_layout()
plt.show()
```
